```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
import statsmodels.formula.api as smf
```

```python
train = pd.read_csv('train_competition_2023.csv')
train = train.groupby('obs').mean()
train.reset_index(inplace=True)
train=train.drop(columns=['obs','sub_id'])
xtrain = train.iloc[:,:-2].copy()
xtrain
```

```python
reg = smf.ols('y_1~num_0+num_1+num_2+C(cat_0)+C(cat_1)+C(cat_2)+C(cat_3)+C(cat_4)+t_0+t_1+t_2+t_3+t_4',data=train).fit()
reg.summary()
```

```python
from sklearn.linear_model import Ridge
```

```python
cv = KFold(n_splits=5, random_state=13, shuffle=True)

def score_given_alpha(alpha, cv, x, y1,y2):
    score1 = []
    score2 = []
    for train_index, val_index in cv.split(x):
        x_train, x_val = x.iloc[train_index,:], x.iloc[val_index,:]
        y1_train, y1_val = y1[train_index], y1[val_index]
        y2_train, y2_val = y2[train_index], y2[val_index]
        reg1 = Ridge(alpha=alpha).fit(x_train, y1_train)
        reg2 = Ridge(alpha=alpha).fit(x_train, y2_train)
        score1.append(mean_absolute_error(y1_val, reg1.predict(x_val)))
        score2.append(mean_absolute_error(y2_val, reg2.predict(x_val)))

    return (np.mean(score1)+np.mean(score2))/2

def find_best_score(x_train, y1,y2, cv,alphas):
    best_alpha = alphas[0]
    best_score = np.inf
    for alpha in alphas:
        score = score_given_alpha(alpha, cv, x_train, y1,y2)
        if score < best_score:
            best_score = score
            best_alpha = alpha
    return best_score, best_alpha
```

```python
best = find_best_score(xtrain,ytrain1,ytrain2,cv,alphas=[0.001,0.01, 0.1, 1, 10])
print('the best regularization constant is:',best[1])
print('MAE returned by the selected model:',best[0])
```
